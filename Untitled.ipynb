```python
#dropped unrequired atributes
df = df.drop(['Artist.Name', 'Loudness..dB..','Valence.','Acousticness..','Unnamed: 0','Danceability','Liveness'], axis = 1)
#new Data Frame
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier


popularity = pd.DataFrame(df['Genre_enc'])
nam_train, nam_test, pop_train, pop_test = train_test_split(df[['Beats.Per.Minute', 'Energy', 'Speechiness.']],
                                                            df['Popularity'],
                                                            test_size=0.3,
                                                            random_state=4, stratify=None)

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(nam_train, pop_train)  # Shikha lav f:X→Y
Hypothesis_prediction_y = knn.predict(nam_test)# Antaz h:X→Y
```
